**Handeling Missing Data**

```python
#Handeling Missing data
#As we can see in deck only 203 values are not null so droping deck column
#Here we are also droping embarked_town as it has missing value and not impact full in dataset
data=data.drop(columns=['deck','embark_town'],axis=1)
data.head()
```

```python
data.info()
```

```python
# let's handle age null values for that we will use replacing values
data['age'].fillna(data['age'].mean(),inplace=True)
data.head()
```

```python
data['embarked'].fillna('s',inplace=True)
# As only 3 values are missing
```

```python
data.isnull().any()
```
